```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
import os
```

```python

# --- Helper functions ---
def load_class_file(path):
    data = np.loadtxt(path)
    return data
```

```python

def stratified_split_per_class(X, y, train_frac=0.7, seed=42):
    rng = np.random.RandomState(seed)
    X_train_list, y_train_list = [], []
    X_test_list, y_test_list = [], []
    classes = np.unique(y)
    for c in classes:
        Xc = X[y==c]
        idx = np.arange(len(Xc))
        rng.shuffle(idx)
        n_train = int(np.floor(train_frac * len(Xc)))
        train_idx, test_idx = idx[:n_train], idx[n_train:]
        X_train_list.append(Xc[train_idx])
        X_test_list.append(Xc[test_idx])
        y_train_list.append(np.full(len(train_idx), c))
        y_test_list.append(np.full(len(test_idx), c))
    X_train = np.vstack(X_train_list)
    y_train = np.concatenate(y_train_list)
    X_test = np.vstack(X_test_list)
    y_test = np.concatenate(y_test_list)
    return X_train, X_test, y_train, y_test
```

```python

def compute_mean_cov(X):
    mu = np.mean(X, axis=0)
    cov = np.cov(X, rowvar=False, bias=True)
    return mu, cov
```

```python

def make_common_sigma2I(class_covs):
    scalar_vars = [np.trace(cov)/cov.shape[0] for cov in class_covs]
    sigma2 = np.mean(scalar_vars)
    Sigma = sigma2 * np.eye(class_covs[0].shape[0])
    return Sigma, sigma2
```

```python

def make_common_fullSigma(class_covs):
    return np.mean(class_covs, axis=0)

def make_diag_per_class(class_covs):
    return [np.diag(np.diag(cov)) for cov in class_covs]
```

```python

def discriminant(x, mu, Sigma, prior):
    x = np.atleast_2d(x)
    D = mu.shape[0]
    try:
        invS = np.linalg.inv(Sigma)
        sign, logdet = np.linalg.slogdet(Sigma)
    except np.linalg.LinAlgError:
        Sigma = Sigma + 1e-6*np.eye(D)
        invS = np.linalg.inv(Sigma)
        sign, logdet = np.linalg.slogdet(Sigma)
    diff = x - mu.reshape(1, -1)
    quad = -0.5 * np.sum(diff @ invS * diff, axis=1)
    return quad - 0.5 * logdet + np.log(prior)
```

```python

def predict_bayes(X, mus, covs, priors):
    K = len(mus)
    scores = np.zeros((X.shape[0], K))
    for k in range(K):
        scores[:, k] = discriminant(X, mus[k], covs[k], priors[k])
    preds = np.argmax(scores, axis=1) + 1
    return preds, scores
```

```python
# Paths to your uploaded dataset files
base = ""
f1 = os.path.join(base, "LS_Group15/LS_Group15_Class1.txt")
f2 = os.path.join(base, "LS_Group15/LS_Group15_Class2.txt")
f3 = os.path.join(base, "LS_Group15/LS_Group15_Class3.txt")

# Load datasets
X1, X2, X3 = load_class_file(f1), load_class_file(f2), load_class_file(f3)

# Labels
y1 = np.ones(X1.shape[0], dtype=int)
y2 = np.ones(X2.shape[0], dtype=int) * 2
y3 = np.ones(X3.shape[0], dtype=int) * 3

# Combine
X = np.vstack([X1, X2, X3])
y = np.concatenate([y1, y2, y3])

print("Shapes:", X1.shape, X2.shape, X3.shape)
print("Total:", X.shape)
```

```python
X_train, X_test, y_train, y_test = stratified_split_per_class(X, y, train_frac=0.7, seed=123)

print("Train size:", X_train.shape, "Test size:", X_test.shape)
print("Train per-class:", {c: sum(y_train==c) for c in np.unique(y_train)})
print("Test per-class:", {c: sum(y_test==c) for c in np.unique(y_test)})
```

```python
classes = np.unique(y_train)
mus, covs = [], []

for c in classes:
    mu, cov = compute_mean_cov(X_train[y_train==c])
    mus.append(mu)
    covs.append(cov)
    print(f"Class {c} mean:\n{mu}\nCov:\n{cov}\n")

mus = np.array(mus)
covs = np.array(covs)
priors = np.array([np.mean(y_train==c) for c in classes])

print("Priors:", priors)
```

```python
# Case 1: σ²I
Sigma_sigma2I, sigma2 = make_common_sigma2I(covs)
covs_case1 = np.array([Sigma_sigma2I for _ in range(len(classes))])
print("Case 1 σ² =", sigma2)

# Case 2: Common full Σ
Sigma_common_full = make_common_fullSigma(covs)
covs_case2 = np.array([Sigma_common_full for _ in range(len(classes))])

# Case 3: Diagonal per class
covs_case3 = np.array(make_diag_per_class(covs))

# Case 4: Full per class
covs_case4 = covs.copy()
```

```python
from sklearn.metrics import classification_report

results = {}
for i, covs_used in enumerate([covs_case1, covs_case2, covs_case3, covs_case4], start=1):
    preds, _ = predict_bayes(X_test, mus, covs_used, priors)
    acc = accuracy_score(y_test, preds)
    cm = confusion_matrix(y_test, preds, labels=[1,2,3])
    results[i] = {'preds': preds, 'acc': acc, 'cm': cm}
    print(f"Case {i}: Accuracy={acc:.4f}")
    print(cm, "\n")
```

```python
buffer = 1.0
xmin, ymin = X[:,0].min()-buffer, X[:,1].min()-buffer
xmax, ymax = X[:,0].max()+buffer, X[:,1].max()+buffer

xx, yy = np.meshgrid(np.linspace(xmin, xmax, 400),
                     np.linspace(ymin, ymax, 400))
grid = np.column_stack([xx.ravel(), yy.ravel()])

cases = [covs_case1, covs_case2, covs_case3, covs_case4]
titles = [
    r"Case 1: $\sigma^2 I$",
    r"Case 2: Common $\Sigma$",
    r"Case 3: Diagonal per class",
    r"Case 4: Full per class"
]

fig, axes = plt.subplots(2, 2, figsize=(14,12))
axes = axes.ravel()

for idx, ax in enumerate(axes, start=1):
    covs_plot = cases[idx-1]
    preds_grid, _ = predict_bayes(grid, mus, covs_plot, priors)
    Z = preds_grid.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.25)
    for c, marker in zip(classes, ['o','s','D']):
        ax.scatter(X_train[y_train==c,0], X_train[y_train==c,1], marker=marker, s=20, label=f"Train C{c}", edgecolor='k')
        ax.scatter(X_test[y_test==c,0], X_test[y_test==c,1], marker=marker, facecolors='none', s=40, label=f"Test C{c}")
    ax.set_title(titles[idx-1])
    ax.legend(fontsize='small')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))
axes = axes.ravel()

for i, ax in enumerate(axes, start=1):
    cm = results[i]['cm']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[1,2,3])
    disp.plot(ax=ax, cmap=plt.cm.Blues, colorbar=False)
    ax.set_title(f"Case {i} (Acc={results[i]['acc']:.3f})")

plt.tight_layout()
plt.show()
```
